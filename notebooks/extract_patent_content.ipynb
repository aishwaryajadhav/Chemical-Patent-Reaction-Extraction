```python
def get_text_own(obj):
	if obj is None: return ''
	else: 		
		return str(obj.get_text().strip())

def extract_content(file_path):
    soup = BeautifulSoup(open(file_path).read(), 'html.parser')
	# content = "TITLE\n" + get_text_own(soup.find('invention-title'))
    content = "TITLE\n" + get_text_own(soup.find('invention-title')) + '\n'
    content += "ABSTRACT\n" + get_text_own(soup.find('abstract')) + '\n'

    soup_des    = soup.find('description')
    if soup_des is not None:
        children_hp = [x for x in soup_des.children if x.__class__.__name__ !='str' and x.name in ['heading', 'p']]
        for child in children_hp:
            if child.name == 'p':
                child_filt   = [x for x in child.children if x.name not in ['tables', 'table']]
                temp_list = [get_text_own(x) if x.__class__.__name__ == 'Tag' else x.strip() for x in child_filt]
                temp_list = [t.strip() for t in temp_list if t.strip()!=""]
                if(len(temp_list) > 0):
                    temp_list = ' '.join(temp_list)
                else:
                    temp_list = ""
            else:
                temp_list = get_text_own(child)
            
            if(temp_list.strip() not in  ["",'\n']):
                content += (temp_list.strip() + '\n')

    claims = [get_text_own(x).replace('\n',' ') for x in soup.find_all('claim')]
    content += "CLAIMS\n" + "\n".join(claims)
 
    return content
```

```python
save_dir = '/home/anjadhav/Chemical-Patent-Reaction-Extraction/data/eval_patent_texts/organic'

def process (patent_paths, save_dir):
    for f in patent_paths:
        try:
            content = extract_content(f)
            file_name = "".join(f.split('/')[-1].split('.')[:-1])
            print(file_name)
            file = open(f'{save_dir}/{file_name}.txt','w')
            file.write(content)
        except Exception as e:
            print(e)
    
```
